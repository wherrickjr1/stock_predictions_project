```python
# List of allowed values
allowed_values = [
'3.0 - Hold',
 '3.1 - Hold',
 '3.2 - Hold',
 '3.3 - Hold',
 '3.4 - Hold',
 '3.5 - Hold',
 '3.6 - Underperform',
 '3.7 - Underperform',
]

# Filter rows
filtered = analyst_ratings[
    analyst_ratings['averageAnalystRating'].isin(allowed_values)
]
```

```python
filtered.head()
```

```python
positive_pct = (filtered['percent_diff'] > 0).mean() * 100
negative_pct = (filtered['percent_diff'] < 0).mean() * 100

print(f"Positive: {positive_pct:.2f}%")
print(f"Negative: {negative_pct:.2f}%")
```

```python
# Function to compute % positive and % negative
def pos_neg_percent(s):
    total = len(s)
    pos = (s > 0).sum() / total * 100
    neg = (s < 0).sum() / total * 100
    return pd.Series({'Positive %': pos, 'Negative %': neg})

# Group by ratings
result = analyst_ratings.groupby('averageAnalystRating')['percent_diff'].apply(pos_neg_percent)

print(result)
```
